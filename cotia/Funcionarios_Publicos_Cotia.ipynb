### Importação de bibliotecas

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
```

```python
fun_cotia = pd.read_csv('funcionariosCotia.csv', sep='|', encoding='iso-8859-1')
```

```python
fun_cotia.head()
```

### Visão geral da remuneração dos funcionários

```python
fun_cotia.Remuneração.describe()
```

### Faz uma pequena, mas não completa ordenação por tempo de casa

```python
fun_cotia.sort_values('Tempo de Casa', inplace=True)
```

### Agrupa os funcioários por secretaria e retorna uma visão geral da remuneração para cada uma delas

```python
fun_cotia.groupby('Secretaria').Remuneração.describe()
```

### Gráfico do numeros e funcionários por secretaria

```python
plt.rcParams['figure.figsize']=(15,10)
sns.countplot(data=fun_cotia, x='Secretaria')
plt.title('Numero de funcionarios por secretaria')
plt.ylabel('Numero de funcionarios')
plt.xticks(rotation=90)
plt.show()
```

### Gráfico que mostra o número de funcionários por tempo de casa
#### Todo:
- Colocar os funcionários com até um ano numa mesma categoria
- Limitar o gráfico para imprimir até 36 anos

```python
plt.rcParams['figure.figsize']=(15,10)
sns.countplot(data=fun_cotia, x='Tempo de Casa')
plt.title('Numero de funcionarios por Tempo de Casa')
plt.ylabel('Numero de funcionarios')
plt.xticks(rotation=90)
plt.show()
```

### Gráfico do valor destinado para pagamento com salários por cada secretaria
#### Todo:
- Fazer uma gráfico separada para educação e saúde e as demais todas juntas

```python
fun_cotia.pivot_table(index='Secretaria', values='Remuneração', aggfunc=np.sum).plot(kind='bar')
plt.title('Total gasto com salario por secretaria')
plt.ylabel('Total de gasto com remuneração')
plt.show()
```

### Gráfico do valor destinado à remuneração por cargo
#### Todo:
- Como são muitos cargos pensar uma outra forma de fazer essa vizualização. Talvez plotar apenas os 10 maiores valores

```python
fun_cotia.pivot_table(index='Cargo', values='Remuneração', aggfunc=np.sum).plot(kind='bar')
plt.title('Total da folha de pagamento por Cargo')
plt.ylabel('Valores em R$')
plt.show()
```

### Distribuição por faixa salarial
#### É possivel observar que a imensa maioria (fazer calculo da porcentagem) está localizada até R$5000,00. Talvez fazer um gráico por faixas salariais?

```python
salarios = fun_cotia.Remuneração
sns.distplot(salarios, bins =32, kde=False)
plt.title('Distribuição da faixa salarial')
plt.ylabel('Numero de Funcionarios')
plt.show()
```

```python
salario_educacao = fun_cotia[fun_cotia['Secretaria'] == 'Secretaria Municipal de Educacao'].Remuneração
```

### Descrição da remuneração da Secretaria da Educação

```python
salario_educacao.describe()
```

### Distribuição por faixa salarial apenas da Secretaria da Educação

```python
sns.distplot(salario_educacao, kde=False, bins=20)
plt.title('Distribuição salarial dos funcionarios da Secretaria de Educação')
plt.ylabel('Numero de Funcionarios')
plt.show()
```
